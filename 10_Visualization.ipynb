```python
result_4s = pd.read_csv('./Result/result_4_nn_run1.csv', index_col=0)
result_9s_gem = pd.read_csv('./Result/result_9_gem_run1.csv', index_col=0)
result_9s_me0 = pd.read_csv('./Result/result_9_me0_run1.csv', index_col=0)
# result_9s_full = pd.read_csv('./Result/result_9_full_nn_run1.csv', index_col=0)
result_9s_full = pd.read_csv('./Result/result_9_full_nn_run2.csv', index_col=0)
result_csctf = pd.read_csv('./Result/results_csctf.csv', index_col=0)
```

```python
points = np.array(range(24, 48)) / 20
bias = []
resolution = []
for i in points:
    temp = result_4s[(result_4s['eta_median']>i-0.05) & (result_4s['eta_median']<i+0.05)]
    bias.append(temp['delta_1/pt'].mean())
    resolution.append(2 * temp['delta_1/pt'].std())
#     resolution.append((temp['delta_1/pt'].max() - temp['delta_1/pt'].min())/2)
```

```python
fig, ax = plt.subplots()
ax.errorbar(points, bias, color='firebrick', yerr=resolution, errorevery=1)
plt.ylim(-0.1, 0.2)
ax.grid(True)
# ax.set_title('all errorbars')
```
